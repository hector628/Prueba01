```python
import math
x=float(input("Indica el valor de x: "))
y=float(input("Indica el valor de y: "))

print("Selecciona la operación que desees realizar: ")
print("1.Suma")
print("2.Resta")
print("3.Multiplicación")
print("4.División")
print("5.Potencia")
print("6.Logaritmo(logaritmo del primer número)")

n=int(input("¿Cuál opción quieres?"))

if n==1:
    resultado=x+y
    print(resultado)
elif n==2:
    resultado=x-y
    print(resultado)
elif n==3:
    resultado=x*y
    print(resultado)
elif n==4 and y!=0:
    resultado=x/y
    print(resultado)
elif n==5:
    resultado=x**y
    print(resultado)
elif n==6 and x>0:
    resultado=math.log10(x)
    print(resultado)
else:
    print("No se puede dividir entre cero. El logaritmo de cero no está definido")
```
